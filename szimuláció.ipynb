```python
def DeltaE(n, npm1):
    return 1/2*(npm1**2-n**2)
```

```python
def SimulationStep(n):
    r = random.random(1)
    direction = ((r-1/2)/abs(r-1/2))
    npm1 = int(n+direction)
    if DeltaE(n, npm1) < 0:
        n = npm1
    else:
        rand = random.random(1)
        if rand < exp(-betaka2*DeltaE(n, npm1)):
            n = npm1
        else:
            n = n
    return (n)
```

```python
temp =0
steps = []
for i in range(1000000):
    n = temp
    steps.append(temp)
    temp = SimulationStep(n)
    
    #print(temp)
steps.append(temp)
print(temp)
```
